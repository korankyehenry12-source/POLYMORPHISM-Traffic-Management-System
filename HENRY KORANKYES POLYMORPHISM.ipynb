```python
class TrafficDevice:
    """Base class for all smart traffic devices."""
    
    def activate(self):
        print("[TrafficDevice] Device is now active")
    
    def deactivate(self):
        print("[TrafficDevice] Device is now inactive")
```

```python
class TrafficLight(TrafficDevice):
    def activate(self):
        print("[TrafficLight] Switching to GREEN -> YELLOW -> RED")


class SpeedCamera(TrafficDevice):
    def activate(self):
        print("[SpeedCamera] Scanning road for speeding vehicles")


class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("[PedestrianSignal] Displaying WALK sign")
```

```python
# Create device objects
light = TrafficLight()
camera = SpeedCamera()
signal = PedestrianSignal()

# Store in a list
all_devices = [light, camera, signal]

# Activate all devices polymorphically
print("--- TRAFFIC CONTROL CENTER ---")
for device in all_devices:
    device.activate()
print("------------------------------")
```

```python
class EmergencySiren(TrafficDevice):
    def activate(self):
        print("[EmergencySiren] ALERT! Flashing red lights and loud siren!")
```

```python
# Add EmergencySiren to the existing list
all_devices.append(EmergencySiren())

# Same loop, now handles 4 devices!
print("--- UPDATED TRAFFIC CONTROL CENTER ---")
for device in all_devices:
    device.activate()
print("--------------------------------------")
```
